### enumerate() 

is a built-in Python function that helps you loop through a list (or any iterable) while also keeping track of the index (position) of each item.

Syntax:
enumerate(iterable, start=0)
iterable → list, tuple, string, etc.
start → (optional) starting index (default = 0)

```python
mylist
```

```python
for i in mylist:              # for loop will print only value not indexes
    print(i)
```

```python
# with enumerate()
```

```python
for i, v in mylist:
    print(i, v)
```

```python
for i in enumerate(mylist):      # here we have tested code without 'v', and o/p comes in tuple
    print(i)
```

```python
for i, v in enumerate(mylist):            # indexing is zero by default
    print(i, v)
```

```python
# custoom start
```

```python
for i, v in enumerate(mylist, start = 2):      # indexing will print from 2
    print(i,v)
```

```python
numbers = [10, 35, 20, 40, 15]

count =0
for i, v in enumerate(numbers):
    if v>25:
        print(f"index.{i} --> {v}")
        count = count +1
print(f"Total count = {count}")
```
