```python
# important dict methods: keys()

empl = {
    'name': 'akshay',
    'age': 21
}

print(empl.keys())      # returns all keys
```

```python
# values()

print(empl.values())     # returns all keys
```

```python
# items: returns all dict items

print(empl.items())
```

```python
# loop in dictionary

empl = {
    'name': 'akshay',
    'age': 21
}
```

```python
for i in empl:             # its for keys
    print(i)
```

```python
# for values

for j in empl.values():
    print(j)
```

```python
# for both key + value

for i, j in empl:        # this will not work
    print(i,j)
```

```python
for i, j in empl.items():        # this will  work
    print(i,j)
```

```python
# finding top selling product: common in DA

sales = {
    'mobile': 500,
    'TV':     297,
    'headphone': 700,
    'tablet': 205,
}

top_product = max(sales, key=sales.get)
print(top_product)
```

```python
# high spending customers

customers = {
    'akshay' : 14500,
    'jay' : 4500,
    'ram' : 3500,
}

high_value = []

for i,j in customers.items():
    if j > 5000:
        high_value.append(i)

print(high_value)
```

```python
# Scenario: Word Frequency Analysis: Very common in NLP and Text Analytics.

txt = "python python sql powerbi sql"

words = txt.split()
print(words)

freq = {}

for i in words:
    freq[i] = freq.get(i,0) + 1

print(freq)
```

```python
# convert list to dict: very common in ETL work

products = ['laptop','mobile','tv']
sales = [10,29,6]

result = dict(products, sales)      # this will not work
```

```python
products = ['laptop','mobile','tv']
sales = [10,29,6]

result = dict(zip(products, sales))

print(result)
```

```python
# group data

cities = ["Pune", "Mumbai", "Pune", "Delhi"]

count = {}

for city in cities:
    count[city] = count.get(city, 0) + 1

print(count)
```
