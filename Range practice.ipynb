```python
# find maximum value manually using range()

sales = [100, 800, 500, 350, 490, 170]

max_val = sales[0]

for i in range(1,len(sales)):
    if sales[i] > max_val:
        max_val = sales[i]

print('Maximum value: ', max_val)
```

```python
# Calculate a 3-day moving average using range().


sales = [100, 120, 130, 150, 170]
moving_averages = []


for i in range(len(sales) - 2):

    three_day_sum = sales[i] + sales[i+1] + sales[i+2]
    avg = three_day_sum / 3
    
    
    moving_averages.append(round(avg, 2))

print("Moving Averages:", moving_averages)

```
