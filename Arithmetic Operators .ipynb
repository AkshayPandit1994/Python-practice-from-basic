Arithmetic operators perform mathematical calculations.

| Operator | Meaning             | Example |
| -------- | ------------------- | ------- |
| +        | Addition            | 10 + 5  |
| -        | Subtraction         | 10 - 5  |
| *        | Multiplication      | 10 * 5  |
| /        | Division            | 10 / 5  |
| //       | Floor Division      | 10 // 3 |
| %        | Modulus (Remainder) | 10 % 3  |
| **       | Power               | 2 ** 3  |

```python
# Scenario 6: Compound Growth

growth_rate = 1.10

future_sales = 100000 * (growth_rate ** 3)

print(future_sales)
```

The Formula BreakdownThe script calculates the Future Value using the standard compound interest/growth formula:

Future Sales= Initial Value * (1+ Growth Rate)t

Initial Base: 100,000
Growth Factor: 1.10 ** 3 = 1.331
Calculation: 100,000 × 1.331 = 133,100

Year-by-Year GrowthYear 0: 100,000
Year 1: 110,000 (+10,000)
Year 2: 121,000 (+11,000)
Year 3: 133,100 (+12,100)
