```python
# 14. Recursive functions: function calling itself

def countdown(n):
    if n==0:
        return
    print(n)
    countdown(n-1)
countdown(4)
```

```python
# 17. function inside function argument

def square(x):
    return x*x

def calculate(func, value):
    return func(value)

print(square(5))

print(calculate(square,5))
```

```python
# 18. map function (very important)

# without function:

numbers =[1,2,3,4]

result =[]

for i in numbers:
    result.append(i*2)
print(result)
```

```python
# with function

def double(x):
    return x*2

result = list(map(double, [1,2,3,4]))
print(result)
```

```python
# 19 . filter function (very important)

def is_high_sale(x):
    return x > 500

sales = [200,800,300,1000]

result = list(filter(is_high_sale, sales))

print(result)
```

```python
# 21. Functions with Dictionaries

def total_revenue(products):
    return sum(products.values())

sales = {
    'Laptop':50000,
    'Mobile':30000
}

print(total_revenue(sales))
```

```python
# 25. Customer Segmentation Function
def customer_type(amount):

    if amount >= 100000:
        return "Premium"

    elif amount >= 50000:
        return "Gold"

    else:
        return "Regular"

print(customer_type(70000))
```

```python
# 26. KPI Calculation Function
def growth_rate(old, new):
    return ((new-old)/old)*100

print(growth_rate(1000,1200))
```

```python
# 27. Professional Data Analyst Function

def sales_summary(sales):

    return {
        "Total Sales": sum(sales),
        "Average Sales": sum(sales)/len(sales),
        "Max Sales": max(sales),
        "Min Sales": min(sales)
    }

print(sales_summary([100,200,300,400]))
```
