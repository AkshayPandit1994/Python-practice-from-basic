```python
# convert string to bytes: common in data analysis:          encode()

txt = 'data analysis'

print(txt)

b = txt.encode()

print(b)
```

```python
txt = 'data analysis'

print(txt)

b = b(txt)

print(b)
```

```python
# convert bytes to string: common in data analysis:          decode()


t = b'data science'

print(t)

t1 = t.decode()

print(t1)
```

```python
# important bytes methods: 1. decode()

data = b'akshay'

print(data.decode())
```

```python
# 2. count()

data = b'python'

print(data.count(b'a'))               # count(b,'a'): b is mandatory. 'a' count 'a' in given string
```
