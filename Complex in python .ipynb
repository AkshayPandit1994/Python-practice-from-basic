```python
# Signal Processing (Very Important)

signal = 4 + 3j

magnitude = abs(signal)

print(magnitude)
print(type(magnitude))


# Here:

# abs() gives signal strength

# This is common in:

# radar systems
# communication systems
# speech analysis
```

```python
# Fourier Transform in Data Science

# In advanced analytics, Fourier Transform converts:

# time-domain data
# → frequency-domain data

# Libraries like:

# NumPy
# SciPy

# return complex values.


import numpy as np

data = [1,2,3,4,5]

result = np.fft.fft(data)

print(result)


# Output contains complex numbers.

# Used in:

# stock trend frequency analysis
# sound wave analysis
# sensor data analysis
```

```python
# Scenario 3: Electrical Data Analysis

# Electrical engineers analyze:

# voltage
# current
# impedance

# using complex numbers.

voltage = 220+30j
current = 55 + 20j

power = voltage * current

print(power)
```
