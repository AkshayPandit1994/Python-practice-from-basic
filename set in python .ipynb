# important set operations

```python
# union : combine all uniqie values

a= {1,2,3}
b= {2,3,4}

a.union(b)
print(a)


# Why this happensunion() creates a new set: The a.union(b) method returns a completely new set containing elements from both sets.a remains unchanged: Because the returned set is never saved to a variable, the original set a is not modified
```

```python
a= {1,2,3}
b= {2,3,4}

a =a.union(b)    # this is how we can fix this
print(a)
```

```python
a= {1,2,3}
b= {2,3,4}

print(a | b)
```

```python
# intersection: common values

a= {1,2,3}
b= {2,3,4}

print(a & b)
```

```python
print(a or b)
```

```python
# diffrence values
a= {1,2,3}
b= {2,3,4}

print(a-b)                 # values in a not in b
print(b-a)                 # values in b not in a
```

```python
# symmetric diffrence: values in either set but not in both

a= {1,2,3}
b= {2,3,4}

print(a^b)
print(b^a)
```

```python
# common skills 

employee1 = {'Python', 'SQL', 'Excel'}

employee2 = {'Python', 'Power BI', 'SQL'}

common_skills = employee1 & employee2

print(common_skills)
```

```python
# missing training: required skils

required = {'python','sql','power_bi'}
employees_skills = {'python','sql'}

missing  = required - employees_skills

print(missing)
```
