### FOR()

```python
flist = ['Python', 100, 16.66, 'data', 9+9j,]
flist
```

```python
for i in flist:
    print(i)
```

```python
name= 'Python'
for i in name:
    print(i)
```

```python
for i in range(15):             # range()
    print(i)
```

```python
for i in range(1,10,2):    # start  with 1, end is 10, which is 10-1=9, gap = gap
    print(i)
```

```python
for i in range(2,10,2):    
    print(i)
```

```python
for i in range(1,15):
    if i ==10:
        print("Available")
```

```python
for i in range(1,15):
    if i ==10:
        print("Available")
    else:
        print('Missing')
```

```python
for i in range(1,15):
    if i ==100:
        print("Available")
        break                                 # we can stop loop using  break()
else:
        print('Missing')
```

```python
if 179 in range(1,15):                # advance
    print('available')
else:
    print('missing')
```

```python
### continue   ---> skip the given step
```

```python
for i in range(1,6):
    if i == 5:
        continue     
    print(i)
```

```python
for i in range():
    print(i)
```

```python
for i in range(1,4):        # Nested for loop
    for j in range(1,4):
        
        print(i,j)
```

```python
for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            print(i,j,k)
```

```python
for i in range(5):
    print(i)
else:
    print('Loop end!')
```

```python
for r in range(3):
    for c in range (3):
        print('*', end= ' ')
    print()
```

```python
for r in range(3):
    for c in range (3):
        print('*', end='\n')       # this will print on new line
    print()
```

```python
for r in range(3):
    for c in range (3):
        print('*', end=' ')       # this will print on same line
    print()
```

```python
for r in range(4):           # rows
    for c in range (5):       # columns
        print('*', end=' ')       # this will print on same line
    print()
```
