```python
# Create a frozenset of departments and validate user input.

departments = frozenset(['operations','it','accounting','claims','endorsement'])

dept = input('Enter department name: ')

for i in departments:
    if i == dept:
        print('Valid entry')
        
    else:
        print('please enter valid name')
```

```python
# another way for same o/p


departments = frozenset(['operations','it','accounting','claims','endorsement'])

dept = input('Enter department name: ').lower()

if dept in departments:
    print('Valid entry!')
else:
    print('Invalid entry')
```

```python
# same program with variation

departments = frozenset(['operations', 'it', 'accounting', 'claims', 'endorsement'])

dept = input('Enter department name: ').lower()

found = False

for i in departments:
    if i == dept:
        found = True
        break

if found:
    print('Valid entry')
else:
    print('Please enter a valid department name')
```

```python
# Compare two shopping baskets.

basket_1 = frozenset(['tv', 'laptop', 'mobile'])
basket_2 = frozenset(['tv', 'mobile', 'earbuds'])

common_products = basket_1.intersection(basket_2)

print(common_products)
```

```python

unique_to_basket_2 = basket_2.difference(basket_1)

print(unique_to_basket_2)
```

```python
# Build a simple recommendation system using frozenset keys.

recommendations = {
    frozenset(['tv','sound system']): 'wall mount',
    frozenset(['mobile','earbuds']): 'power bank',
    frozenset(['laptop','mouse']): 'keyboard',
}

customer_basket = frozenset(['tv','sound system'])

if customer_basket in recommendations:
    print('Recommended products: ', recommendations[customer_basket])
else:
    print('No recommendation')
```

```python
# another example

recommendations = {
    frozenset(['python', 'sql']): 'power bi',
    frozenset(['excel', 'sql']): 'python',
    frozenset(['python', 'power bi']): 'tableau'
}

skills = frozenset(['python', 'sql'])

print(
    "Recommended skill:",
    recommendations.get(skills, 'No recommendation')
)
```
