```python
# Practical Data Analysis Examples, 1. Count Byte Frequencies

data = bytearray([1,1,1,2,3,2,2,2,3,3,3])

freq = {}

for byte in data:
    freq[byte] = freq.get(byte,0) + 1

print(freq)
```

```python
# 2. Remove Invalid Sensor Values


sensor_data = bytearray([10,20,255,30,255,40])

clean_data = bytearray(x for x in sensor_data if x != 255)

print(clean_data)
```
