Important String Methods for Data Analysts

| Method         | Use                  |
| -------------- | -------------------- |
| `lower()`      | standardize text     |
| `upper()`      | convert to uppercase |
| `strip()`      | remove spaces        |
| `replace()`    | fix errors           |
| `split()`      | separate values      |
| `join()`       | combine values       |
| `find()`       | locate text          |
| `count()`      | frequency analysis   |
| `startswith()` | pattern checking     |
| `endswith()`   | file validation      |

```python
# professional method

review = input('Enter your review: ').lower()

negative_words = [
    'late',
    'poor',
    'disgusting',
    'bad service',
    'behaviour issue'
]

if any(word in review for word in negative_words):
    print('Negative review')
else:
    print('Positive review')
```
