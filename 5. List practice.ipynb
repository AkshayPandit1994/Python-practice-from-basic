```python
marks = [70,75,80,65,90]
```

```python
marks.append(85)                # it will add single number/value
marks.extend([60,95])           # # it will add multiple number/value
marks
```

```python
marks.insert(2,77)            # insert number at 2 nd position
marks
```

```python
marks.remove(65)                  # remove specific gicen value
marks
```

```python
marks.pop()                    # remove last one
marks
```

```python
marks[0]=72                     # Update a mark
marks
```

```python
if 90 in marks:          # Check if a mark exists
    print("Yes")
```

```python
print("Total marks:", len(marks))             # find total student
```

```python
marks.sort()                         # ascending order
marks
```

```python
marks.reverse()                      # reverse
marks
```

```python
print('Final List:')

for i,v in enumerate(marks, start=1):
    print(f"{i}.{v}")
```

```python
print('Marks greater than 80:')

for i in marks:
    if i>80:
        print(i)
```
