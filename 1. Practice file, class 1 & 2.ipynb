# repr: 

```python
print(chr(65))            # o/p diffrent 
```

```python
print(repr(chr(65)))         # # o/p diffrent when we put repr 


# Notice

# quotes appear
# Python shows the object's representation
```

```python
print(chr(10))             # blank line, because character 10 is newline.
```

```python
print(repr(chr(10)))


# Much better!

# You can actually see the newline character.
```

```python
# another esample

print(chr(9))         # tab
```

```python
print(repr(chr(9)))            # Now we know it's a Tab.       
```
