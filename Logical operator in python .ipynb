Logical operators are used to combine multiple conditions and return either True or False.

| Operator | Meaning                             | Example                              |
| -------- | ----------------------------------- | ------------------------------------ |
| `and`    | Both conditions must be True        | `age > 18 and salary > 50000`        |
| `or`     | At least one condition must be True | `city == "Pune" or city == "Mumbai"` |
| `not`    | Reverses the result                 | `not is_active`                      |


```python
# 1.and operator: returns true when all conditions are true.

sales = 5000
profit = 2000

print(sales > 4000 and profit > 1900)
```

```python
# data analyssi scenario: high performing products

sales = 150000
rating  = 4.7

if sales > 100000 and rating >4:
    print('Top poducts')
```

```python
# 2. or operator: Returns True if at least one condition is True.

runs = 5000
avg = 49.5

print(runs > 6000 or avg > 45)
```

```python
# Data Analysis Scenario: Marketing Target

customer_spend = 12000
visit = 2

if customer_spend >10000 or visit >5:
    print('Target customer!')
```

```python
# 3. not opertor: reverse the result

is_active = False

print(not is_active)
```

```python
is_active = True

print(not is_active)
```
