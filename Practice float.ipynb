```python
# salary increment system

salary = 95000.91

increment = salary * 0.10

revised_salary = salary + increment

print(revised_salary)
print(round(revised_salary,1))
```

```python
# discount system

price = 2000.91
discount = 10

discount_amount = price * discount / 100

final_price = price - discount_amount

print(round(final_price, 1))
```

```python
# restaurant bill

bill = 1000
gst = 18

gst_bill = bill*gst / 100

final_bill= bill+ gst_bill

print(final_bill)
```

```python
# currency conversion

usd = 95.64
rate = 81.11

inr = usd * rate

print(round(inr,2))
```
