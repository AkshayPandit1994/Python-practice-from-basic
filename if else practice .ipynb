```python
# Classify age as Child, Adult, Senior.

age = int(input('Enter your age:'))

if age < 18:
    category= 'Minor or child'
elif age >18 and age <60:
    category = 'Adult'
else:
    category = 'Senior'

print(category)
```

```python
# another way

age = int(input('Enter your age: '))

if age < 18:
    category = 'Minor or Child'
elif 18 <= age < 60:
    category = 'Adult'
else:
    category = 'Senior'

print(category)
```

```python
# Detect duplicate names.

names = ['akshay', 'rahul', 'priya', 'rahul']

name = input('Enter name: ').lower()

if names.count(name) > 1:
    print('Duplicate name')
else:
    print('Unique name')
```
