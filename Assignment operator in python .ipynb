```python
# Important Data Analysis Scenarios, Scenario 1: Running Total

sales = [100,200,300]

total = 0

for i in sales:
    total += i         # total= total + i
print(total)
```

```python
# Scenario 2: count records

cust_name = ['a','b','c','d','e']

count  = 0

for i in cust_name:
    count += 1                        # logic changes here, count = count + 1

print(count)
```

```python
# Scenario 2: building frequency dictionary

products = ['TV', 'TV', 'AC', 'TV']

freq = {}

for i in products:
    freq[i]= freq.get(i,0)+1
print(freq)
```

```python
# Scenario 4: Updating Revenue

revenue = 100000

revenue += 5000
revenue -= 2000

print(revenue)
```

```python
# Scenario 5: Incrementing Loop Variable

day = 1

while day <=5:
    print(day)
    day += 1
```
