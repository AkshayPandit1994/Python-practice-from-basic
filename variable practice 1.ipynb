```python
template = "{} is my country, all {} are my brothers and sisters."
result = template.format('India','Indians')                             
print(result)
```

```python
# indexed fomatting(reusing values)

# {0} always grabs the 1st item ("India"), {1} grabs the 2nd ("Indians")
template = "{0} is my country. I love {0}, and all {1} are my family."

result = template.format("India", "Indians")
print(result)
```

```python
# 3. Named Formatting (Keyword Arguments)

template = "{country} is my home, and all {people} are my family."

result = template.format(people="Indians", country="India")
print(result)
```

```python
# Modern Alternative: F-Strings (Python 3.6+)

country = 'India'
people = 'Indians'

result = f'{country} is my country, all {people} are my brothers and sisters'
print(result)
```
