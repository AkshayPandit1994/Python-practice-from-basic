```python
# In Python, Boolean represents True or False. It is mainly used for conditions and decision making. Boolean is a data type in Python that represents True or False and is used in conditions and logical operations.

# Comparison Operators (Return Boolean)
#| Operator | Meaning          |
#| -------- | ---------------- |
#| `==`     | Equal            |
#| `!=`     | Not equal        |
#| `>`      | Greater          |
#| `<`      | Less             |
#| `>=`     | Greater or equal |
#| `<=`     | Less or equal    |

# Boolean Summary

#| Concept | Meaning      |
#| ------- | ------------ |
#| `True`  | Yes / 1      |
#| `False` | No / 0       |
#| `and`   | Both true    |
#| `or`    | Any one true |
#| `not`   | Reverse      |
```
