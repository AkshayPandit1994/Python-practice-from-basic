```python
# find total sales

sales = {
    "Laptop": 50000,
    "Mobile": 30000,
    "Tablet": 15000,
    "TV": 45000
}

total = sum(sales.values())
print(total)
```

```python
# Find highest selling product.

sales = {
    "Laptop": 50000,
    "Mobile": 30000,
    "Tablet": 15000,
    "TV": 45000
}

highest = max(sales, key=sales.get)

print('highest selling product : ', highest)
print('sales', sales[highest])
```

```python
# Find lowest selling product.

sales = {
    "Laptop": 50000,
    "Mobile": 30000,
    "Tablet": 15000,
    "TV": 45000
}

Lowest = min(sales, key=sales.get)

print('lowest selling product : ', Lowest)
print('sales', sales[Lowest])
```

```python
# Calculate average sales.

sales = {
    "Laptop": 50000,
    "Mobile": 30000,
    "Tablet": 15000,
    "TV": 45000
}

avg = sum(sales.values()) / len(sales)

print(avg)
```

```python
# Display products with sales > 30000

sales = {
    "Laptop": 50000,
    "Mobile": 30000,
    "Tablet": 15000,
    "TV": 45000
}

for i,j in sales.items():
    if j >= 30000:
        print(i, ':', j)
```

```python
# another version

for product, amount in sales.items():
    if amount > 30000:
        print(product, ":", amount)
```

```python
# add new product

sales = {
    "Laptop": 50000,
    "Mobile": 30000,
    "Tablet": 15000,
    "TV": 45000
}

sales['AC'] = 4000
print(sales)
```

```python
# update existing sales

sales = {
    "Laptop": 50000,
    "Mobile": 30000,
    "Tablet": 15000,
    "TV": 45000
}

sales['Mobile'] = 32000
print(sales)
```

```python
# remove a product

sales = {
    "Laptop": 50000,
    "Mobile": 30000,
    "Tablet": 15000,
    "TV": 45000
}

sales.pop('Tablet')

print(sales)
```

```python
# sort by sales high to low

sales = {
    "Laptop": 50000,
    "Mobile": 30000,
    "Tablet": 15000,
    "TV": 45000
}

sorted_sales = dict(
    sorted(
        sales.items(),
        key=lambda item: item[1],
        reverse=True
    )
)

print(sorted_sales)
```

```python
#Convert results into another dictionary.

sales = {
    "Laptop": 50000,
    "Mobile": 30000,
    "Tablet": 15000,
    "TV": 45000
}

summary = {
    "Total Sales": sum(sales.values()),
    "Average Sales": sum(sales.values()) / len(sales),
    "Highest Product": max(sales, key=sales.get),
    "Lowest Product": min(sales, key=sales.get)
}

print(summary)
```

```python
# dashboard

sales = {
    "Laptop": 50000,
    "Mobile": 30000,
    "Tablet": 15000,
    "TV": 45000
}

dashboard = {
    "Total Sales": sum(sales.values()),
    "Average Sales": sum(sales.values()) / len(sales),
    "Highest Product": max(sales, key=sales.get),
    "Lowest Product": min(sales, key=sales.get),
    "Products Above 30000":
        [p for p, s in sales.items() if s > 30000]
}

print(dashboard)
```
