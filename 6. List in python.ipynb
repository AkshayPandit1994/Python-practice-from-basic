Python list for Data Analysis

In Python, a list is one of the most important data types for data analysis because it stores multiple values in one variable.

A list can contain:

numbers
strings
mixed data
duplicate values
even other lists


Why Lists Are Important in Data Analysis

Lists are used for:

storing sales records
customer names
marks
temperatures
product prices
transaction amounts
survey answers
machine learning data preparation

```python
# creating list

marks = [88,96,75,69,89]

print(marks)
print(type(marks))
```

```python
# access data from list

marks = [88,96,75,69,89,96,90]

print(marks[0])
print(marks[1])
print(marks[-1])
```

```python
# list slicing

profit = [12,14,19,23,25,5,8,11]

print(profit[2:])
print(profit[:])
print(profit[3:6])
print(profit[:9])

print(profit[-4:-1])
print(profit[::-1])
```

```python
profit = [12,14,19,23,25,5,8,11]

print(profit[-6:-1])       # both are skipping right side element
print(profit[1:6])         # both are skipping right side element
```

```python
# updating list value

lst = [10,20,30,50]

lst.index(50) 

lst[3] = 40

print(lst)    # list changes
```

```python
# adding data

menu = ['burger','pizza','momos']

menu.append('sandwich')
print(menu)
```

```python
# extend

data = [10,20,30,40]
print(data)

data.extend([50,60,70])

print(data)
```

```python
data.extend
```

```python
# insert

height = [5,4,6,7,4]

height.insert(0,9)

print(height)
```

```python
# remove data

card = ['debit','credit','voter','visiting']

card.remove('visiting')   # remove by value

print(card)
```

```python
# pop: remove value by index

days = [21,22,33,23,24]

days.pop(2)
print(days)
```

```python
days.pop()          # default removes last value
print(days)
```

```python
# clear

Petrol_rate = [100,85,99,65]
print(Petrol_rate)

Petrol_rate.clear()              # removes and return  blank string
print(Petrol_rate)
```

```python
# length of list

numbers= [14,25,36,85,74,95,15,35] 

print(len(numbers))               # it gives total number(1 to 8), indexing start from 0-7
```

```python
# finding maximum and minimum

numbers = [45,63,88,2,44,77,2,11,99,6,88,100]

print(max(numbers))
print(min(numbers))
```

```python
# sum & average

data = [100,5004,3603,841]

total = sum(data)
print(total)

# average = sum(data) / len (data)         # both are same
average = total / len(data)
print(average)
```

```python
# sorting data

numbers= [1,2,3,4,5,6,8,5,6,9,4,7,1,2,3,5,4,8,3,2,5,4,7,1,5,6,9]

numbers.sort()
print(numbers)

numbers.sort(reverse=True)
print(numbers)
```

```python
# checking data existence

products = ['remote','keyboard','lamp','mouse','charger']

'remote' in products            # both lines works

print('remote' in products)
```

```python
# counting values

hotel_names = ['Akshay_inn','Hayat','FourX','Taj', 'Taj']

print(hotel_names.count('Taj'))         # return occurance
```

```python
# finding position

cities = ['Pune', 'Mumbai', 'Delhi']

print(cities.index('Delhi'))
```

```python
# looping through list

sales = [1000,5600,3614,1254] 

for i in sales:
    print(i)
```

```python
# conditional analysis

sales = [1000,5600,3614,1254] 

for i in sales:
    if i > 1500:
        print('Top sell amountL:', i)
```

```python
# list comprehension

price = [1000,700,900,100,1100]

tax_amount = [i * 0.18 for i in price]

print('Tax amount: ', tax_amount)
```

```python
price = [1000,700,900,100,1100]

tax_amount = [i * 0.18 for i in price]

print('Tax amount: ', tax_amount)

final_amount = [p + t for p, t in zip(price, tax_amount)]

print('Final amount: ', final_amount)
```

```python
# filtering data

sales = [500, 2000, 7000, 300]

high_sales = [i for i in sales if i >1000]
print('High sale:',high_sales)

low_sales = [i for i in sales if i <1000]
print('Low sale:', low_sales)
```

```python
# nested list: table -like data

dish_list= [
    ['dosa',80],
    ['panipuri',20],
    ['paav bhaji',80],
    ['misal',70]
]

print(dish_list[0])
print(dish_list[1])
print(dish_list[2])
print(dish_list[3])


print('\n\n',dish_list[0][0])
print(dish_list[0][1])
```

```python
# real data analysis, mini scenario

sales = [500,400,900,300,750,890,950,150,640]

print('Total sales',sum(sales))
print('Minimum sale',min(sales))
print('Maximum sale',max(sales))
print('Average of sale',sum(sales) / len(sales))
```

```python
# student result analysis

marks = [35, 75, 90, 20, 88]

for i in marks:
    if i >=35:
        print(i,'Pass')
    else:
        print(i,'Fail')
```

```python
# Expense Tracker

expenses = [100,600,900,450,350]

budget = 2500
print(sum(expenses))

if sum(expenses) > budget:
    print('Bugget amount exceed')
else:
    print('Under budget')
```

```python
# advanced data analysis: remove null values

data = [10, None, 20, None, 30]
print(data)

clean = [i for i in data if i is not None]  # remove None
clean1 = [i for i in data if i is None]     # remove all other values except None

print(clean)
print(clean1)
```

```python
# convert string to integer

data = ['10', '20', '30']

numbers = [int(i) for i in data]
numbers1 = [i for i in data]

print(numbers)
print(type(numbers))

print(numbers1)
print(type(numbers1))
```

```python
# Detect outliers

sales = [100,200,300,400,5000,500]

for i in sales:
    if i > 1000:
        print(i, 'Possible outlier, please check!')
```
