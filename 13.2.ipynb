```python
t = (1,2,3,4,5)
ret = ()

for i in range(len(t),-1):
    ret = ret + 1

print(ret)
```

```python
t = (1,2,3,4,5)
ret = ()

for i in range(len(t)-1, -1,-1,-1):
    ret += (t[i])

print(ret)
```

```python
t = (1,2,3,4,5)
ret = ()               # Since tuples are immutable, we cannot modify them — so we create a new one.

for i in range(len(t)-1, -1,-1):
    ret += (t[i],)

print(ret)
```
