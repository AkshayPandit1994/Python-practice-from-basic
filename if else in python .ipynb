```python
# 5. logical operators : AND

sales = 2000
profit = 1100

if sales > 1500 and profit >1000:
    print('Business on good track')
```

```python
# OR  : either 1 condition must true

sales = 2000
profit = 1100

if sales >2500 or profit >1000:
    print('Healthy business')
```

```python
# NOT : reverse condition

active = False

if not active:
    print('Inactive customer')
```

```python
# checking membership

department = 'IT'

if 'IT' in department:
    print('IT department')
else:
    print('Non IT')
```

```python
# another 

dept = input('Enter your Department: ').lower()

allowed_dept = ['it','admin','account','operation']            # list must match

for i in allowed_dept:
    if i == dept:
        print('Allowed')
        break
else:
    print('not allowed')
```

```python
# 10. profit loss analysis

revenue  = 100000
cost = 75000

profit = revenue - cost

if profit > 0:
    print(profit,'profit')
elif profit <0:
    print(profit,'loss')
else:
    print('No gain/loss')
```

```python
# 14. nested if

sales = 12000
profit = 3000

if sales > 10000:
    if profit  > 2500:
        print('Profit')
```

```python
# 15. ternary operator" short form of if else

sales = 6000

status = 'Good' if sales > 5000 else 'Poor'
print(status)
```

```python
# 16. loop + if (most important) analyst often use this

sales = [1000,2000,5000,4000]

for i in sales:
    if i > 2000:
        print('Good sales: ', i)
```

```python
# 17 .list compregension + if : common in DA

sales = [1000,2000,5000,4000]

high_sales= [s for s in sales if s >= 2000]

print(high_sales)
```
