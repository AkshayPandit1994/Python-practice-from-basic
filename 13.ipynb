```python
# example with marks

marks= 5

if marks >= 95:
    print('--> Grade A')
elif marks >=80:
    print('--> Grade B')
elif marks >= 70:
    print('--> grade C')
elif marks >= 50:
    print('--> Passed')
else:
    print('Failed, please try hard')
```

```python
# example with marks with input

marks= int(input('Please enter your marks to know your grade:'))

if marks >= 95:
    print('--> Grade A')
elif marks >=80:
    print('--> Grade B')
elif marks >= 70:
    print('--> grade C')
elif marks >= 50:
    print('--> Passed')
else:
    print('Failed, please try hard')
```

```python
# signal

signal = 'Green'

if signal == 'Red':
    print('STOP')
elif signal == 'Yellow':
    print('Slow down')
elif signal == 'Green':
    print('Go! happy journey')
else:
    print('Please enter valid color')
```

```python
# signal with input

signal = input('Please enter signal:')

if signal == 'Red':
    print('STOP')
elif signal == 'Yellow':
    print('Slow down')
elif signal == 'Green':
    print('Go! happy journey')
else:
    print('Please enter valid color')
```
