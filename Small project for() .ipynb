# Small project based on for()

```python
user1 = int(input('Please enter marks for student 1:').strip())
user2 = int(input('Please enter marks for student 2:').strip())
user3 = int(input('Please enter marks for student 3:').strip())
user4 = int(input('Please enter marks for student 4:').strip())
user5 = int(input('Please enter marks for student 5:').strip())

marks= [user1, user2, user3, user4, user5]

high = marks[0]

for m in marks:

    if m > high:
        high = m

    if m >= 90:
        print(m,'Grade --> A')
    elif m >= 70:
        print(m,'Grade --> B')
    elif m >= 50:
        print(m,'Grade --> C')
    else:
        print(m, 'Failed')

print('HIghest marks:', high)
```
