```python
# float precison problem


######## sometime python gives strange decimal o/p

print(0.1 + 0.2)                             


# it happens because computers store decimal numbers in binary format. it happend only with 0.1 + 0.2, other decimal shows correct value, solution on this is round()
```

```python
 print(round(0.1+0.2))
```

Useful Float Functions

| Function | Use              |
| -------- | ---------------- |
| float()  | Convert to float |
| round()  | Round decimals   |
| abs()    | Absolute value   |
| max()    | Largest value    |
| min()    | Smallest value   |
| sum()    | Total            |

Summary: 

| Concept             | Meaning                         |
| ------------------- | ------------------------------- |
| float               | Decimal number                  |
| Used in             | Analytics, finance, percentages |
| Conversion          | float()                         |
| Rounding            | round()                         |
| Common issue        | Precision problem               |
| Important libraries | NumPy, Pandas                   |
