```python
# find common customers between 2 stores

store_a = {'a','b','c','d','e','f'}
store_b = {'a','b','c','g','h','i'}
    
common_cust = store_a.intersection(store_b)

print(common_cust)
```

```python
# Find duplicate names in a list.

names = ['akshay', 'rahul', 'priya', 'rahul', 'akshay', 'amit']

duplicate = set()

for i in names:
    if names.count(i) > 1:
        duplicate.add(i)

print('Duplicate names: ', list(duplicate))
```

```python
# merge 2 customer database

cust_1 = {'akshay', 'rahul', 'priya'}
cust_2 = {'rahul', 'akshay', 'amit'}

merge = cust_1.union(cust_2)
print(merge)
```

```python
# find students enrolled in both cources

student1= {'python','java','sql'}
student2= {'c','c++','python'}


same = student1.intersection(student2)     # same cource
print(same)

same_enroll= student1 - student2           # diffrent ource
print(same_enroll)
```

```python
# Find employees lacking required skills.

required = {'Python', 'SQL', 'Power BI'}
employee = {'Python', 'SQL'}

missing_skill = required - employee
print(missing_skill)
```

```python
emails = [
    'akshay@gmail.com',
    'rahul@yahoo.com',
    'priya@gmail.com',
    'amit@outlook.com',
    'neha@yahoo.com'
]

domains = set()

for i in emails:
    domain = i.split('@')[1]
    domains.add(domain)

print(domains)
```
