```python
# 9. best practice: 'with' statement

# instead of:

file = open('sample.txt','r')

file.close()
```

```python
# we can use below:

with open('sample.txt','r') as file:
    data =file.read()
print(data)


# Benefits:

# Auto closes file
# Safer
# Professional coding style
```
