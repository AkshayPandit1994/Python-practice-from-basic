# 2. for loop

```python
# scanerio 1: iterate through sales data

sales = [10,20,30,40,50]

for amount in sales:
    print(amount)
```

```python
# scanerio 2: calculate total revenue

sales = [100,500,800,400,300,100]

total = 0

for i in sales:
    total += i

print(total)
```

```python
# scanerio 3: find high value transaction

sales = [22,66,8,11,2,4,99,20,36,55]

for amount in sales:
    if amount > 50:
        print(amount)
```

# 3. while loop

```python
# scanario: 1- count record

count = 0

while count <=10:
    count += 1              # just beacause we add this line first before print, coutn will go till 11 instead of 10
    print(count)
```

```python
count = 0

while count <=10:
    print(count)
    count += 1
```

```python
# scanario: 2- data validation

password = ''

while password == '':
    password = input('Enter password: ')

print('Access granted')
```

```python
# scanario: 3- reading data until end

record_remain = 5

while record_remain >=0:
    print('Procesing record')
    record_remain -= 1
```

# 4. nested loop()

```python
# scenario 1 - product customer analysis

products = ['laptop','mobile']

customers = ['a','b']

for product in products:
    for customer in customers:
        print(product, customer)
        print('\n',customer,product)
  
```

```python
# scenario 2 - creating cross join: useful in reoport combinations

regions = ['north','south']
years = [2024,2025]

for region in regions:
    for year in years:
        print(region, year)
```

# 6. continue

```python
# scenario 1 - ignore missing values:

data = [10,20,30, None,40,50,None,60,70]

for i in data:
    if i == None:
        continue
    print(i)
```

```python
data = [10,20,30, None,40,50,None,60,70]

for value in data:
    if value is None:
        continue
    print(value)
```

```python
# scenario 2- process only active customers

customers = ['active', 'inactive', 'active']

for status in customers:
    if status == 'inactive':
        continue
    print(status)
    print('Proceed')
```

```python
# 4. duplicate detection

names = ['Ram', 'Shyam', 'Ram', 'Akshay']

seen = set()

for name in names:

    if name in seen:
        print("Duplicate:", name)

    seen.add(name)
```
