```python
# why is memoryview useful? with bytearray:

data = bytearray(b'Hello There!')

copy = data[:7]       # python creates new copy
print(copy)
```

```python
# with memoryview:

data = bytearray(b'Hello there')

view= memoryview(data)
print(view)


# No data is copied.

# This saves:

# Memory
# Processing time
# CPU resources
```

```python
# Scenario 1: Access Large Binary Files Efficiently. Imagine a 2 GB data file.

#Without memoryview:

chunk = data[0:1000000]

#A new copy is created.

# With memoryview:

view = memoryview(data)

chunk = view[0:1000000]

# No copy.

# Useful for:

# ETL pipelines
# Large CSV processing
# Data engineering
```

```python
# scenario: 2 modify data without copying

data= bytearray(b'hello')

view = memoryview(data)

view[0]= ord('H')     # string changes

print(data)
print(view)
```

```python
# scenario: 2 working with sensor data

sensor_data = bytearray([10,20,30,40,50])

view = memoryview(sensor_data)

print(view[4])


# Useful in:

# IoT analytics
# Streaming data
# Real-time monitoring
```

```python
#scenario 4: processing image data, images are stored in bytes

pic = bytearray([255,30,15,78])
print(pic)
v= memoryview(pic)

v[0] =200
print(pic)

# Used in:

# OpenCV
# Computer Vision
# Image preprocessing
```

```python
#scenario 5: working with network packets

packet = bytearray(b'akshay')

view = memoryview(packet)

header = view[:2]

print(header)
print(bytes(header))

# Useful for:

# APIs
# Socket programming
# Streaming data
```

```python
# Scenario 6: Large Dataset Slicing, suppose

d =bytearray(range(100))        # normal slicing
print(d)

part = d[10:20]     # it creates copy
print('\nExtracted part:',part)

# with memory view

view = memoryview(d)     # no copy
part1 = view[10:20]

print('\n With momoryview:', part1)
print(bytearray(part1))
```

```python
# Scenario 7: reading binary records

record = bytearray([1,2,3,4,5,6,7,8])
print('Bytearray: ',record)

view= memoryview(record)

first_record=view[:5]

print('\nwith list:',list(first_record))

# Useful in:

# Binary databases
# Log processing
# Data ingestion
```

```python
# scenario 8:  streaming data pipelines

stream = bytearray(range(50))

view = memoryview(stream)

chunk1= view[:10]
chunk2= view[10:20]

print(chunk1)
print(chunk2)

print(list(chunk1))
print(list(chunk2))

print(bytearray(chunk1))
print(bytearray(chunk2))


# Useful for:

# Kafka consumers
# Spark pipelines
# Real-time analytics
```

```python
# Scenario 9: NumPy Performance

# memoryview works nicely with buffer-supporting objects.

import numpy as np

arr = np.array([10,20,30,40])

view = memoryview(arr)

print(view[0])

# Useful for:

# Scientific computing
# Large arrays
# Data science
```

```python
# Scenario 10: Memory Optimization

# Without memoryview:

big_data = bytearray(10000000)

copy = big_data[:]

#Two huge objects exist.

# With memoryview:

big_data = bytearray(10000000)

view = memoryview(big_data)

#Only one object's data is stored.
```

#### Important methods

```python
# tolist()

data = bytearray([10,20,30])

view = memoryview(data)

print(view.tolist())     # convert to list
```

```python
# bytes(): convert view bacl to bytes

data = bytearray(b'ABC')

view = memoryview(data)

print(bytes(view))
```

```python
# nbytes(): total bytes

data = bytearray(b'Python')

view = memoryview(data)

print(view.nbytes)
```

```python
# readonly() Check if data can be modified.

data = b'Python'

view = memoryview(data)

print(view.readonly)
```

```python
# shape:useful with numpy

import numpy as np

arr = np.array([[1,2],[3,4]])

view = memoryview(arr)

print(view.shape)
```

```python
# Data Analysis Practice Examples, Example 1: Average Sensor Reading

sensor = bytearray([25,30,35,40])

view = memoryview(sensor)

avg = sum(view)/len(view)

print(avg)
```

```python
# Example 2: Detect Corrupted Values

data = bytearray([10,20,255,30])

view = memoryview(data)

for value in view:
    if value == 255:
        print("Corrupted")
```

```python
# Example 3: Analyze Packet Header

packet = bytearray(b'HEADDATA')

view = memoryview(packet)

header = bytes(view[:4])

print(header)
```
