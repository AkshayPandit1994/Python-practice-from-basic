Important frozenset Methods

| Method                 | Purpose                |
| ---------------------- | ---------------------- |
| union()                | Combine values         |
| intersection()         | Common values          |
| difference()           | Missing values         |
| symmetric_difference() | Different values       |
| issubset()             | Check subset           |
| issuperset()           | Check superset         |
| isdisjoint()           | Check no common values |
| copy()                 | Copy frozenset         |


```python
a = frozenset([1,2,3])
b = frozenset([3,4,5])

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
```

Data Analyst Interview Questions: Q1. Difference between set and frozenset?

| set                      | frozenset             |
| ------------------------ | --------------------- |
| Mutable                  | Immutable             |
| Can add/remove items     | Cannot modify         |
| Not hashable             | Hashable              |
| Cannot be dictionary key | Can be dictionary key |


What is "Hashable" in Python?

A hashable object is an object that has a fixed value and can be converted into a unique hash number by Python.

Python uses these hash values to store data efficiently in:

Dictionaries (dict)
Sets (set)
Frozensets (frozenset)

Why is frozenset Hashable?

A frozenset cannot be modified after creation.

```python
name = 'akshay'

print(hash(name))
```

```python
num = [1,2,3]      #list nt hashable, beacause list can change, If the value changes, the hash would also need to change, which Python doesn't allow.

print(hash(num))
```

Quick rule to remeber

| Data Type   | Mutable? | Hashable? |
| ----------- | -------- | --------- |
| `int`       | No       | ✅ Yes     |
| `float`     | No       | ✅ Yes     |
| `str`       | No       | ✅ Yes     |
| `bool`      | No       | ✅ Yes     |
| `tuple`*    | No       | ✅ Yes     |
| `list`      | Yes      | ❌ No      |
| `set`       | Yes      | ❌ No      |
| `dict`      | Yes      | ❌ No      |
| `frozenset` | No       | ✅ Yes     |

*A tuple is hashable only if all its elements are hashable.

Easy Memory Trick

Hashable = Immutable (mostly) = Can be used as a dictionary key.
