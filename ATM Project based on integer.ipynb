```python
balance = int(input('Enter bank balance: '))

while True:

    print("\n ----->> ATM Menu <<-----")
    print('1. Check Balance')
    print('2. Deposite Money')
    print('3. Withdraw Money')
    print('4. Exit')


    choice = int(input('Enter choice: '))


    # if choice is to check balance

    if choice ==1:
        print('Current balance=', balance)
        

    # if cust want to deposite money

    elif choice ==2:

        amount = int(input('Enter amount wish to deposite: '))

        balance = balance + amount

        print('Amount deposited successfulluy')

        print('Balance after deposite: ', balance)


    # if cuatomer want to withdraw amount

    elif choice == 3:


        amount = int(input('Enter amount wish to withdraw: '))

        if amount<= balance:

            balance = balance-amount

            print('Please collect your cash: ')
            print('Total balance after amount:, ', balance)
        
        else:

            print('Insufficient balance')



        # if customer wants to exit

    elif choice == 4:

        print('Thank you using Akshay Bank Service, Have a nice day!')

        break


    else:

        print('Invalid choice, please add correct option!')
        
    
```
