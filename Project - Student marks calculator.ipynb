```python
marathi = int(input('Enter marks obtained in Marathi: '))
science = int(input('Enter marks obtained in Science: '))
math = int(input('Enter marks obtained in Math: '))
english = int(input('Enter marks obtained in Emglish: '))


total = marathi + science + math + english

percentage = total / 4

print('\n************ RESULT ************')

print('Marathi: ', marathi)
print('Science: ', science)
print('Maths: ', math)
print('English: ', english)

print('Total: ', total)
print('Percentage: ', percentage)


if marathi < 35 or science < 35 or math < 35 or english < 35:

    print('Result: Fail, try again!')

else:

    print('Result: Pass!')

if percentage >= 92:

    print('Grade= A+')
elif percentage >= 80:

    print('Grade= A')

elif percentage >= 70:

    print('Grade = B')

else:
    print('Grade = C')
        
```
