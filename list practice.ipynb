```python
# list comprehension

price = [921,513,650,129,90,753]

tax = [i * 0.18 for i in price]

print(tax)
```

```python
# nested lists: list in list

data = [10,20,30,[5,15,25],40,50]

print(data)
```

```python
print(data[0])
print(data[1])
print(data[2])
print('\n',data[3])
```

```python
# loops in list

marks = [25,63,79,80,44,50,90]

for i in marks:
    if i >= 40:
        print(i,'Pass')
    else:
        print(i, 'Fail, try hard again!')
```

```python
# filtering data

numbers = [2000,3000,1500,1000,500,1100,450]

high_sales = [i for i in numbers if i >=2000]
low_sales = [i for i in numbers if i <=2000]

print(high_sales)
print(low_sales)
```
