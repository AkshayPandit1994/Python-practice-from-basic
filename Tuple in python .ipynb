Tuple vs list (very important)

| Feature          | Tuple      | List          |
| ---------------- | ---------- | ------------- |
| Symbol           | `()`       | `[]`          |
| Mutable          | No         | Yes           |
| Faster           | Yes        | Slower        |
| Memory Efficient | Yes        | Less          |
| Use Case         | Fixed data | Changing data |


When to Use Tuple in Data Analysis

Use tuple when:

✅ Data should not change
✅ Working with database rows
✅ Returning multiple values
✅ Coordinates or dimensions
✅ Fast iteration needed
✅ Dictionary keys required

```python
# Tuple Comprehension?

# There is NO tuple comprehension directly.

# This creates generator object:

data = (x*x for x in range(5))

print(data)
```

```python
# correct tuple creation

data = tuple(x*x for x in range(5))

print(data)
```
