```python
# Reverse an integer.

# logic behind: Modulus % Gets last digit. and Floor Division // Removes last digit.

rev = int(input('Enter number to reverse: '))

reverse = 0

while rev > 0:
    digit = rev%10
    reverse = reverse * 10 + digit
    rev= rev //10
print('Reverse number: ', reverse)
```

```python
# Palindrome integer in python

# a palindrome number is a number that remain the same after reversing. ex 11, 151 etc.

n = 121

original = n
reverse = 0

while n >0:
    digit = n%10
    reverse = reverse * 10 +digit
    n= n//10

if original == reverse:
    print('Palindrom number')
else:
    print('Not Palindrom')
```

```python
n = int(input('Enter a number: '))

original = n
reverse = 0

while n >0:
    digit = n%10
    reverse = reverse * 10 +digit
    n= n//10

if original == reverse:
    print('Palindrom number')
else:
    print('Not Palindrom')
```
