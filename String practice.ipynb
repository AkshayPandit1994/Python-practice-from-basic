```python
# Replace old city names 

# replace ols city name like Bombay, Madras, Calcutta to Mumbai, Chennai, kolkata

city = input('Enter city name to know new name: ')

if city = 'Bambay' or 'Madras' or 'Calcuta':
    print('Mumbai','Chennai','Kolkata')
else:
    print('City is with update name')
```

```python
# Better Professional Approach (Dictionary)

city = input('Enter city: ')

city_m = {
    'Bombay' : 'Mumbai',
    'Madras' : 'Chennai',
    'Calcuta' : 'Kolkata'
}

if city in city_m:
    print(city_m[city])
else:
    print('No new name found')
```

```python
city = input("Enter city: ").title().strip()

city_mapping = {
    "Bombay": "Mumbai",
    "Madras": "Chennai",
    "Calcutta": "Kolkata"
}

print(city_mapping.get(city, city))
```
